```python
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import autocorrelation_plot
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
#Probably the implemantation is WRONG!
#Initialize parameters and generate 1000 random points from N(0,1) g works from 0 up to 1.2
s=2
g=0.5
positions = [0]
random_p = np.random.normal(0,1,1000)
```

```python
#positions
for r in random_p:
    mv = -1 if r<0 else 1
    value = positions[-1]*(1-g) + s*mv
    positions.append(value)
```

```python
plt.plot(positions)
```

```python
autocorrelation_plot(positions)
```

```python
#calc differences
diff1 = []
#calc differences with absolute values
diff2 = []

for i in range(1,len(positions)):
    value = positions[i] - positions[i-1]
    diff1.append(value)
for i in range(1,len(positions)):
    value = np.abs(positions[i] - positions[i-1])
    diff2.append(value)
```

```python
plt.plot(diff1,'ro')
```

```python
plt.plot(diff2,'bo')
#calculate s
#s = E[diff2]
np.mean(diff2)
```

```python
#g = sqrt(var(diff)/var(positions))
np.sqrt(np.var(diff2)/np.var(positions))
```
