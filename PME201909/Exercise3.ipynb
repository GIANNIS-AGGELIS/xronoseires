```python
import math
import statistics
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 5000
```

# Weiner Process

```python
a = [0] # χρονοσειρα με 1 στοιχειο μονο, η αρχικη θεση ειναι 0
s = 15    # παραγοντας που αλλαζει τη διασπορα
for x in range(1,N):  #για  1-10000 τιμες
    a.append(a[-1] + s*norm.rvs()) #dx=x+x1-γ*t α(-1) ειναι η τελευταια θεση που ειχα
round(statistics.mean(a)) #μεση τιμη 
round(statistics.variance(a))   #διασπορα
plt.plot(a)
```

```python
d = []
for x in range(1,N):
    d.append(a[x]-a[x-1])
print(f"σ = {round(math.sqrt(statistics.variance(d)), 2)}")
```

# Ornstein Process

```python
s = 3          # volatility coefficient
gamma = 0.002  # reversion rate
m = 15         # long-term mean
ornstein = [0]
for x in range(1,N):
    ornstein.append(ornstein[-1] + gamma*(m - ornstein[-1]) + s*norm.rvs())

plt.figure(dpi=200)
plt.title("Ornstein")
plt.plot(ornstein)
statistics.mean(ornstein)
```
